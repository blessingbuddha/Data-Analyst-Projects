```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data
df = pd.read_csv('./EV_pop_dens_ratio_total_cl.csv')


df
```

```python
#removing commas to graph 
df['Total EVs (2)'].replace(',','',regex=True, inplace=True)
df['Total EVs (2)'] = pd.to_numeric(df['Total EVs (2)'])

df
```

```python
# df = pd.read_csv('Data-Analyst-Projects\EV-PROJECT-main\Data Sets\data set 5~alt_fuel_stations(Nov-3-2023).csv', low_memory=False)
```

```python
df['Total Ports (1)'].replace(',','',regex=True,inplace=True)
df['Total Ports (1)'] = pd.to_numeric(df['Total Ports (1)'])

df['Population'].replace(',','',regex=True,inplace=True)
df['Population'] = pd.to_numeric(df['Population'])

# Drop unused columns
df.drop(columns=[
    'Level 1 Ports (1)',
    'Level 2 Ports (1)',
    'DCFC Ports (1)'],
    inplace=True)

# Rename Columns
df.rename(columns={'Total EVs (2)': 'Total EVs'}, inplace=True)
df.rename(columns={'Total Ports (1)': 'Total Ports'}, inplace=True)
df.rename(columns={'Ratio: EVs to Charger Ports': 'EVs : Chargers'}, inplace=True)

df
```

```python
# Sort Total EVs 
df.sort_values('Total EVs', ascending=True, inplace=True)
df
```

```python
# Graph State to Number of EVs

# Sort State column ascending
# Adjust rows to use in graph
stateShort = df['State'][::]
evShort = df['Total EVs'][::]
portShort =df['Total Ports'][::]

# Assign x and y axis
x = stateShort
y = evShort
a = portShort

# Label and format graph
plt.figure(figsize=(10, 10))
plt.yticks([5000, 10000, 25000, 50000, 100000, 200000, 300000, 4000000, 500000])
plt.xlabel('Total EVs')
plt.ylabel('State')
plt.grid('minor')
plt.xticks(rotation='vertical')
plt.title('EVs per State / Ports per State')

# Plot lines
plt.plot(x, y, 'k^--', label = 'EVs')
plt.plot(x, a, 'r^--', label = 'Ports')

# Add legend
plt.legend()

# Plot chart EVs per State / Ports per State
plt.show()
```
